```python
df_country=pd.read_excel('country-code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on = 'Country Code', how='left')
final_df.head()
```

```python
final_df.columns
```

```python
final_df.Country.value_counts() # from this we find how many entry in this country
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3], autopct="%0.2f%%") # here [:3] used to get top 3 countrys
```

observation : Zomato max records or transaction are from India, the US and then UK

```python
ratings =final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:'Rating Count '})
```

```python
ratings
```

```python
ratings.columns = ratings.columns.str.strip()  #Sometimes, column names have extra spaces. Fix them like this
```

## observation
1. When rating is between 4.5 to 4.9 ---> Excellent
2. When rating is between 4.0 to 3.4 ---> Very good
3. When rating is between 3.5 to 3.9 ---> Good
4. When rating is between 2.4 to 3.4 ---> Average
5. When rating is between 1.8 to 2.4 ---> poor

```python
plt.figure(figsize=(15,6))
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings, hue="Rating color",
           palette=['Blue','Red','Orange','Yellow','Green','Green'])
plt.show()
```

## observation
1. not rated count is very high
2. the most number of rating between 2.5 to 3.4

```python
 sns.countplot(x="Rating color",data = ratings, palette=['Blue','Red','Orange','Yellow','Green','Green'])
```
